````python
import requests
from bs4 import BeautifulSoup

def web_scraping_text(url):
  headers = {'User-Agent' : 'Mozilla/5.0'}
  response = requests.get(url, headers=headers)
  if response.status_code == 200:
    page_content = response.content
    soup = BeautifulSoup(response.text, 'html.parser')
    # want to scrape all the text inside the paragraph tags <p>
    paragraphs = soup.find_all('p')
    # Extract text from the paragraph tags
    scraped_data = [p.get_text() for p in paragraphs]
    return scraped_data
  else:
    print("Failed to retieve the webpage with {0}".format(url))
    return []
````

````python
import urllib.parse
import json

command = "outlook /c ipm.note /m recipient1@example.com;recipient@example.com?subject=test&body=hello"

# Parse the command string
parsed = urllib.parse.urlsplit(command)

# Extract the email recipients
recipients = parsed.path.split(";")

# Extract the email subject and body from the query string
query_params = urllib.parse.parse_qs(parsed.query)
subject = query_params.get("subject", [""])[0]
body = query_params.get("body", [""])[0]

# Create a dictionary with the parsed data
data = {
    "command": parsed.scheme,
    "recipients": recipients,
    "subject": subject,
    "body": body
}

# Convert the data to JSON
json_data = json.dumps(data)

print(json_data)
````

````python
import win32com.client as win32

def send_email(subject, body, recipients):
    outlook = win32.Dispatch("Outlook.Application")
    mail = outlook.CreateItem(0)  # 0 represents the MailItem type

    # Set email properties
    mail.Subject = subject
    mail.Body = body
    mail.To = ";".join(recipients)  # Multiple recipients can be separated by semicolon (;)

    # Send the email
    mail.Send()

# Example usage
subject = "Test Email"
body = "This is a test email sent from Python using Outlook."
recipients = ["recipient1@example.com", "recipient2@example.com"]

send_email(subject, body, recipients)
````
